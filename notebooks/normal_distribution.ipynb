# Normal Distribution

Normal distribution is one of the most important probability distributions
in statistics and machine learning.

It is a symmetrical bell-shaped distribution where:
- Mean = Median = Mode
- Most values lie near the center
- Extreme values are rare

Applications:
- Exam scores
- Human height
- Financial analysis
- Machine Learning assumptions

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Liabraries Imported!!")
```

```python
np.random.seed(42)

data = np.random.normal(
    loc=50,
    scale=10,
    size=1000
)
```

```python
print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Standard Deviation:", np.std(data))
print("Minimum:", np.min(data))
print("Maximum:", np.max(data))
```

```python
plt.figure(figsize=(10,7))

sns.histplot(data, kde=True)

plt.title("Normal Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.show()
```

```python
mean = np.mean(data)
std = np.std(data)

plt.figure(figsize=(12,6))

sns.histplot(data, kde=True)

plt.axvline(mean, linestyle="--", label="Mean")
plt.axvline(mean + std, linestyle=":")
plt.axvline(mean - std, linestyle=":")

plt.legend()
plt.show()
```

```python
within_1_std = np.mean((data > mean - std) & (data < mean + std))

print(within_1_std * 100)
```

```python
import scipy.stats as stats

stats.probplot(data, dist="norm", plot=plt)

plt.show()
```

```python
from scipy.stats import skew, kurtosis

print(skew(data))
print(kurtosis(data))
```
